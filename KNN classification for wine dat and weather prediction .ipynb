```python
weather=['Sunny','Sunny','Overcast','Rainy','Rainy','Rainy',
         'Overcast','Sunny','Sunny','Rainy','Sunny',
         'Overcast','Overcast','Rainy']
```

```python
temp=['Hot','Hot','Hot','Mild','Cool','Cool',
      'Cool','Mild','Cool','Mild','Mild','Mild','Hot','Mild']
```

```python
play=['No','No','Yes','Yes','Yes','No','Yes',
      'No','Yes','Yes','Yes','Yes','Yes','No']
```

```python
# Import LabelEncoder
from sklearn import preprocessing
```

```python
le = preprocessing.LabelEncoder()
```

```python
# Converting string labels into numbers.
le = preprocessing.LabelEncoder()
```

```python
# Converting string labels into numbers.
temp_encoded=le.fit_transform(temp)
print(temp_encoded)
```

```python
weather_encoded=le.fit_transform(weather)
print(weather_encoded)
```

```python
play_encoded=le.fit_transform(play)
print(play_encoded)
```

```python
#combining weather and temp into single
X=list(zip(weather_encoded,temp_encoded))
y=le.fit_transform(play)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size = 0.3, 
                                                    random_state = 1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import matplotlib.pyplot as plt
error_rate = []

for i in range(1,10):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    pred = knn.predict(X_test)
    error_rate.append(np.mean(pred != y_test))

plt.figure(figsize=(15,10))
plt.plot(range(1,10),error_rate, marker='o', markersize=7)
plt.show()
```

```python
model = KNeighborsClassifier(n_neighbors=3) #hyper_tuning

# Train the model using the training sets
model.fit(X,y)
model.score(X, y)

#Predict Output
#predicted= model.predict([[0,2]]) # 0:Overcast, 2:Mild
#print(predicted)
```

```python
#Predict Output
predicted= model.predict([[2,1]]) # 0:Overcast, 2:Mild
print(predicted)
```

```python
#Import scikit-learn dataset library
from sklearn import datasets

#Load dataset
wine = datasets.load_wine()
```

```python
# print the names of the features
print(wine.feature_names)
```

```python
# print the label species(class_0, class_1, class_2)
print(wine.target_names)
```

```python
# print the wine data (top 5 records)
print(wine.data[0:5])
```

```python
# print the wine labels (0:Class_0, 1:Class_1, 2:Class_3)
print(wine.target)
```

```python
# print data(feature)shape
print(wine.data.shape)
```

```python
# Import train_test_split function
from sklearn.model_selection import train_test_split
```

```python
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(wine.data, 
                                                    wine.target, 
                                                    test_size=0.3,
                                                    random_state=0) 
```

```python
#Import knearest neighbors Classifier model
from sklearn.neighbors import KNeighborsClassifier

#Create KNN Classifier
knn = KNeighborsClassifier(n_neighbors=7)

#Train the model using the training sets
knn.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = knn.predict(X_test)
```

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
